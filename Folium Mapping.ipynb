```python
import pandas as pd
import folium
```

```python
df = pd.read_csv(r"C:\Python\ob\Folium Mapping\ob_welltypes.csv",
                 usecols=['WELL_ID', 'WELL_TYPE', 'LATDD', 'LONGDD'])
```

```python
#RENAIMING COLUMNS
df.columns = ['BRACS Well ID', 'Well Type', 'Latitude', 'Longitude']
```

```python
df
```

```python
#DISPLAY BLANK FOLIUM MAP
map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], width=750, height=500,
                 zoom_start=8, 
                 control_scale=True)
map
```

```python
#ADDING MARKERS
folium.Marker(location=[df.Latitude.mean(), df.Longitude.mean()]).add_to(map)
map
```

```python
map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()],
                zoom_start=8, control_scale=True)

folium.Marker(location=[df.Latitude.mean(), df.Longitude.mean()], 
              icon=folium.Icon(color='red', icon='info', prefix='fa')).add_to(map)
              
map
```

```python
map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], 
                 zoom_start=8, control_scale=True)

#Loop through each row in the dataframe

for i,row in df.iterrows():
    #Setup the content of the popup
    iframe = folium.IFrame('BRACS Well ID:' + str(row["BRACS Well ID"]))
    
    #Initialise the popup using the iframe
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    
    #Add each row to the map
    folium.Marker(location=[row['Latitude'],row['Longitude']],
                  icon=folium.Icon(color='red', icon='info', prefix='fa'),
                  popup = popup, c=row['Well Type']).add_to(map)
map
```

```python
# Set up the colours based on well's purpose
purpose_colour = {'Oil or Gas':'red', 'oil or gas':'red', 'Withdrawal of Water':'blue'}

map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], 
                 zoom_start=8, control_scale=True)

#Loop through each row in the dataframe
for i,row in df.iterrows():
    #Setup the content of the popup
    iframe = folium.IFrame(f'BRACS Well ID: {str(row["BRACS Well ID"])} \n Well Type: {str(row["Well Type"])}')
    
    #Initialise the popup using the iframe
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    
    try:
        icon_color = purpose_colour[row['Well Type']]
    except:
        #Catch nans
        icon_color = 'gray'
    
    #Add each row to the map
    folium.Marker(location=[row['Latitude'],row['Longitude']],
                  popup = popup, 
                  icon=folium.Icon(color=icon_color, icon='info', prefix='fa')).add_to(map) 
                   
map.save("my_map_well_density_Washington.html")
map
```
